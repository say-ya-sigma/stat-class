```python
import json
import matplotlib.pyplot as plt

with open('partyData.json') as f:
    partyData = json.load(f)

totalFreq = partyData[-1][1]

def genLabel(line):
    return "{n} {p}%".format(n=line[0], p=str(round(line[1] / totalFreq * 100, 1)))

labels = list(map(genLabel, partyData[:-1]))
freq = list(map(lambda x: x[1],partyData[:-1]))

patches, text = plt.pie(freq, startangle=90, counterclock=False)
plt.legend(patches, labels, bbox_to_anchor=(0, -0.1), loc='upper left', borderaxespad=0)
plt.show()
```
