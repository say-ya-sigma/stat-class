```python
import json
import matplotlib.pyplot as plt

with open('partyData.json') as f:
    partyData = json.load(f)

table = plt.table(
    cellText=partyData,
    colLabels=['supporting party', 'freq.'],
    colWidths=[0.4, 0.1],
    cellLoc='center',
    loc='center'
)

table.set_fontsize(25)
table.scale(3, 3)

plt.axis('off')

plt.show()
```

```python
import json
import matplotlib.pyplot as plt

with open('partyData.json') as f:
    partyData = json.load(f)

totalFreq = partyData[-1][1]

print('n = {total}'.format(total=totalFreq))


def calcRadio(line):
    new = [
        line[0],
        str(round(line[1] / totalFreq * 100, 1)) + '%'
    ]
    return new


partyDataRatio = list(map(calcRadio, partyData[:-1]))

table = plt.table(
    cellText=partyDataRatio,
    colLabels=['supporting party', 'radio'],
    colWidths=[0.4, 0.1],
    cellLoc='center',
    loc='center'
)

table.set_fontsize(25)
table.scale(3, 3)

plt.axis('off')

plt.show()
```
